```python
ham=df[df['label']=='ham']
spam=df[df['label']=='spam']
```

```python
ham=ham.sample(spam.shape[0])
```

```python
data = pd.concat([ham, spam], ignore_index=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(data['message'],data['label'],test_size=0.3,random_state=0,shuffle=True)
```

# 2. **Building The Model(Random Forest)**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

# Use Pipeline (capital P) to import the class
from sklearn.pipeline import Pipeline
```

```python
classifier = Pipeline([("tfidf",TfidfVectorizer()),("classifier",RandomForestClassifier(n_estimators=10))]) # Capitalize 'P' in Pipeline
```

```python
classifier.fit(x_train,y_train)
```

## **3.Predicting The Results (Random Forest)**

```python
y_pred = classifier.predict(x_test)
```

```python
y_test,y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
0.9309576837416481*449
```

```python
confusion_matrix(y_test,y_pred)
```

```python
 print (classification_report(y_test,y_pred))
```

# **4. Building The Model(SVM)**

```python
from sklearn.svm import SVC
```

```python
# Use SVC instead of the overwritten 'svm' variable
svm = Pipeline([("tfidf",TfidfVectorizer()),("classifier",SVC(C = 100,gamma='auto'))])
```

```python
svm.fit(x_train,y_train)
```

# **5.Predicting The results (SVM)**

```python
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
